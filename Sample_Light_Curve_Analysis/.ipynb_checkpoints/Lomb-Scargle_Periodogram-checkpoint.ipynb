# Lomb-Scargle Periodogram Tutorial
https://docs.astropy.org/en/stable/timeseries/lombscargle.html

The Lomb-Scargle periodogram method produces a frequency spectrum of a time series signal, akin to Fourier Analysis, though it utilizes a least squares regression to fit sinusoidal functions to the data. It can be used to find the frequency of a periodic signal, or the absence of one, in the time series. The primary advantage of the Lomb-Scargle periodogram over Fourier analysis is that it does not require evenly spaced time sampling.

````python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from astropy.timeseries import LombScargle
from astropy.timeseries import TimeSeries
from astropy.time import Time
from astropy import units as u
````

#### First create noisy sinusoidal measurements with random time sampling (with frequency 1)

````python
rand = np.random.default_rng(42)
t = 100 * rand.random(100)
y = np.sin(2 * np.pi * t) + 0.1 * rand.standard_normal(100)
dy = 0.1 * (1 - rand.random(100))
````

````python
fig = plt.figure(figsize=(8,5), dpi=100)
plt.errorbar(t, y, yerr = dy, fmt='.')
plt.title('Noisy Sinusoidal Data')
plt.xlabel('Time')
plt.ylabel('Y')
````

#### Fit with Lomb-Scargle Method

The ```autopower()``` method automatically picks a frequency space to perform the least squares regression, with the maximum frequency being five times the 'average Nyquist frequency'. We can alter this by changing the ```nyquist_factor```, or by specifying a frequence space we want to use.

````python
frequency_auto, power_auto = LombScargle(t, y, dy).autopower()
frequency_1, power_1 = LombScargle(t, y, dy).autopower(nyquist_factor=1)
frequency_15, power_15 = LombScargle(t, y, dy).autopower(nyquist_factor=15)

frequency = np.linspace(1e-3, 5, 10000)
power = LombScargle(t, y, dy).power(frequency)
````

````python
fig, ax = plt.subplots(2, 2, figsize=[8,4], sharey=True, dpi=150)

ax[0,0].plot(frequency_auto, power_auto, label='Autopower')
ax[0,1].plot(frequency_1, power_1, label='Nyquist Factor = 1')
ax[1,0].plot(frequency_15, power_15, label='Nyquist Factor = 15')
ax[1,1].plot(frequency, power, label='Custom Frequency Space')

plt.subplots_adjust(left=0.1,
                    bottom=0.1, 
                    right=0.9, 
                    top=0.9, 
                    wspace=0.2, 
                    hspace=0.4)

ax[0,0].set_ylabel('Lomb-Scargle Power', fontsize=8)
ax[1,0].set_ylabel('Lomb-Scargle Power', fontsize=8)
fig.suptitle('Lomb-Scargle Periodograms')
for axis in ax.ravel():
    axis.set_xlabel('Frequency', fontsize=8)
    axis.tick_params(axis='both', which='major', labelsize=8)
    axis.legend(prop={'size': 8})
````

As expected, there is a peak at a frequency of 1 cycle per unit time. When our maximum frequency is too low (i.e. in the periodogram using a Nyquist factor of 1), we see that the peak is absent and no periodicity is found.

# Time Series Example

Now we explore the `TimeSeries` class in astropy, which allows us to fold the data at the frequency we get from the periodogram.

#### Again create a noisy sinusoidal measurement 

````python
rand2 = np.random.default_rng(10)
t2 = 100 * rand2.random(100) + 2458849
y2 = np.sin(2 * np.pi * t2) + 0.1 * rand2.standard_normal(100) + 2
dy2 = np.ones(len(y2))*0.1
````

````python
fig = plt.figure(figsize=(8,5), dpi=100)
plt.errorbar(t2, y2, yerr = dy2, fmt='.')
plt.title('Noisy Time Series Data at Random Sampling')
plt.xlabel('JD')
plt.ylabel('Flux (mJy)')
````

#### Then convert the data to a `TimeSeries` object and perform the Lomb-Scargle periodogram

````python
times = Time(t2, format='jd')
ts = TimeSeries(time = times, data={'flux': y2 * u.mJy, 'flux_error' : dy2 * u.mJy})
````

Using `autopower()`, we can also specify the minimum and maximum frequencies explicitly, making sure that the units are consistent.

````python
frequency_ts, power_ts = LombScargle(ts.time, ts['flux'], dy=ts['flux_error']).autopower(minimum_frequency=0.1/u.d, maximum_frequency=1.9/u.d)
````

````python
fig = plt.figure(figsize=(10,4), dpi=100)
plt.title('Lomb-Scargle Periodogram')
plt.plot(frequency2, power2)
plt.xlabel('Frequency (Cycles per Day)')
plt.ylabel('Power')
````

#### Extract the peak frequency and fold

````python
best_freq = frequency_ts[np.argmax(power_ts)]
ts_folded = ts.fold(period=1./best_freq)
````

````python
fig = plt.figure(figsize=(10,4), dpi=100)
plt.errorbar(ts_folded['time'].jd, ts_folded['flux'], yerr=ts_folded['flux_error'], fmt='.')
plt.title('Data Folded at Peak Frequency')
plt.xlabel('Phase')
plt.ylabel('Flux (mJy)')
````

Of course, this is a near-perfect signal, so we can easily fit a sinusoid to the folded data. Next, we repeat this analysis on a real light curve of a short timescale variable star from Gaia.

# Gaia Light Curve Analysis

The script `lightCurve.py` contains a function for retrieving single Gaia light curves, as well as several other useful functions for dealing with Gaia light curves that will be used below.

````python
import sys
sys.path.append('..')
import lightCurve
````

#### Get Light Curve from Gaia

We pick a short-timescale variable star with a clear periodicity

````python
source_id = 376526416902123392 # short-timescale variable star
product_tb = lightCurve.retrieveLC(source_id)
````

#### Light Curve Analysis

Get the light curves for each band individually and plot

````python
gbandLC = lightCurve.getLC(product_tb, source_id, band = 'G')
rpbandLC = lightCurve.getLC(product_tb, source_id, band = 'RP')
bpbandLC = lightCurve.getLC(product_tb, source_id, band = 'BP')
````

The `plotLC` function is designed for when more than one light curve needs to be plotted, but it can be used to plot a single light curve as well. The light curve inputs must be dictionaries with the source id as keys and the light curves as the items. We do not need the star `SkyCoord` object or Gaia table row since we are not plotting the ellipsoid crossing time on this light curve.

````python
lightCurve.plotLC(str(source_id), None, None, {str(source_id) : gbandLC}, {str(source_id) : bpbandLC}, {str(source_id) : rpbandLC}, median=False, normalize=True, time=None, cross=False)
````

Create `TimeSeries` objects for each band light curve

````python
gtime, rtime, btime = Time(gbandLC['time'] + 2455197.5, format='jd'), Time(rpbandLC['time'] + 2455197.5, format='jd'), Time(bpbandLC['time'] + 2455197.5, format='jd')
gflux, rflux, bflux = np.array(gbandLC['flux']/np.median(gbandLC['flux'])), np.array(rpbandLC['flux']/np.median(rpbandLC['flux'])), np.array(bpbandLC['flux']/np.median(bpbandLC['flux']))
gflux_err, rflux_err, bflux_err = np.array(gbandLC['flux_error']/np.median(gbandLC['flux'])), np.array(rpbandLC['flux_error']/np.median(rpbandLC['flux'])), np.array(bpbandLC['flux_error']/np.median(bpbandLC['flux']))
````

````python
gts = TimeSeries(time = gtime, data={'flux' : gflux, 'flux_error' : gflux_err})
rts = TimeSeries(time = rtime, data={'flux' : rflux, 'flux_error' : rflux_err})
bts = TimeSeries(time = btime, data={'flux' : bflux, 'flux_error' : bflux_err})
````

Perform the Lomb-Scargle periodogram

````python
gfreq, gpower = LombScargle(gts.time, gts['flux'], gts['flux_error']).autopower(minimum_frequency=1/u.d,
                                                   maximum_frequency=10/u.d)
rfreq, rpower = LombScargle(rts.time, rts['flux'], rts['flux_error']).autopower(minimum_frequency=1/u.d,
                                                   maximum_frequency=10/u.d)
bfreq, bpower = LombScargle(bts.time, bts['flux'], bts['flux_error']).autopower(minimum_frequency=1/u.d,
                                                   maximum_frequency=10/u.d)

fig, ax = plt.subplots(3, figsize=[8,5], sharex=True, dpi=150)

ax[0].plot(gfreq, gpower, color='g', label='G')
ax[1].plot(rfreq, rpower, color='r', label='RP')
ax[2].plot(bfreq, bpower, color='b', label='BP')
ax[0].set_title('Lomb-Scargle Periodogram')
ax[2].set_xlabel('Frequency (Cycles per Day)')
ax[1].set_ylabel('Power')

for axis in ax.ravel():
    axis.tick_params(axis='both', which='major', labelsize=8)
    axis.legend(prop={'size': 8})
````

````python
gbestfreq = gfreq[np.argmax(gpower)]
rbestfreq = rfreq[np.argmax(rpower)]
bbestfreq = bfreq[np.argmax(bpower)]

print(f'G band period: {1/gbestfreq}')
print(f'RP band period: {1/rbestfreq}')
print(f'BP band period: {1/bbestfreq}')
````

**This is the same period found in https://arxiv.org/pdf/2206.05591.pdf (p. 10)**

Now fold the data at the extracted frequency

````python
gts_folded = gts.fold(period=1./gbestfreq, normalize_phase=True, epoch_phase=0.25)
rts_folded = rts.fold(period=1./rbestfreq, normalize_phase=True, epoch_phase=0.25)
bts_folded = bts.fold(period=1./bbestfreq, normalize_phase=True, epoch_phase=0.25)
````

````python
fig = plt.figure(figsize=(6,3), dpi=150)

plt.errorbar(gts_folded['time'], gts_folded['flux'], yerr=gts_folded['flux_error'], fmt = 'g.')
plt.errorbar(rts_folded['time'], rts_folded['flux'], yerr=rts_folded['flux_error'], fmt = 'r.')
plt.errorbar(bts_folded['time'], bts_folded['flux'], yerr=bts_folded['flux_error'], fmt = 'b.')
plt.title('Light Curves Folded at Peak Frequency')
plt.xlabel('Phase')
plt.ylabel('Normalized Flux')
````

This is also a relatively ideal light curve from Gaia; many periodic signals are not as cleanly extracted, and most variable stars do not display purely sinusoidal variations. However, even in these cases, the Lomb-Scargle periodogram can be used to find the frequency of a periodic signal, and folding the light curve at that frequency gives us much more information about the star. For examples of this analysis on non-sinusoidal variables, see the RR Lyrae and Eclipsing Binary notebooks.
