# Eclipsing Binary Periodogram

Although eclipsing binary systems do not show a sinusoidal variation in flux, their variability still has a well-defined period that can be extracted with the Lomb-Scargle periodogram. The light curve folded at the frequency can generally be fit with a double Gaussian function.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from astropy.timeseries import TimeSeries, LombScargle
from astropy.time import Time
from astropy import units as u
```

```python
from astroquery.gaia import Gaia
from astropy.table import Table, QTable

import sys
sys.path.append('..')
import lightCurve
```

#### Get Light Curve from Gaia

We pick a known eclipsing binary from the Gaia archive

```python
source_id = 5276980236450011264 # eclipsing binary
product_tb = lightCurve.retrieveLC(source_id)
```

### Light Curve Analysis

Get the light curves for each band individually and plot

```python
gbandLC = lightCurve.getLC(product_tb, source_id, band = 'G')
rpbandLC = lightCurve.getLC(product_tb, source_id, band = 'RP')
bpbandLC = lightCurve.getLC(product_tb, source_id, band = 'BP')
```

The `plotLC` function is designed for when more than one light curve needs to be plotted, but it can be used to plot a single light curve as well. The light curve inputs must be dictionaries with the source id as keys and the light curves as the items. We do not need the star `SkyCoord` object or Gaia table row since we are not plotting the ellipsoid crossing time on this light curve.

```python
lightCurve.plotLC(str(source_id), None, None, {str(source_id) : gbandLC}, {str(source_id) : bpbandLC}, {str(source_id) : rpbandLC}, median=False, normalize=True, time=None, cross=False)
```

Create `TimeSeries` objects for each band light curve

```python
gtime, rtime, btime = Time(gbandLC['time'] + 2455197.5, format='jd'), Time(rpbandLC['time'] + 2455197.5, format='jd'), Time(bpbandLC['time'] + 2455197.5, format='jd')
gflux, rflux, bflux = np.array(gbandLC['flux']/np.median(gbandLC['flux'])), np.array(rpbandLC['flux']/np.median(rpbandLC['flux'])), np.array(bpbandLC['flux']/np.median(bpbandLC['flux']))
gflux_err, rflux_err, bflux_err = np.array(gbandLC['flux_error']/np.median(gbandLC['flux'])), np.array(rpbandLC['flux_error']/np.median(rpbandLC['flux'])), np.array(bpbandLC['flux_error']/np.median(bpbandLC['flux']))
```

```python
gts = TimeSeries(time = gtime, data={'flux' : gflux, 'flux_error' : gflux_err})
rts = TimeSeries(time = rtime, data={'flux' : rflux, 'flux_error' : rflux_err})
bts = TimeSeries(time = btime, data={'flux' : bflux, 'flux_error' : bflux_err})
```

Perform the Lomb-Scargle periodogram

```python
gfreq, gpower = LombScargle(gts.time, gts['flux'], gts['flux_error']).autopower(minimum_frequency=1/u.d,
                                                   maximum_frequency=5/u.d)
rfreq, rpower = LombScargle(rts.time, rts['flux'], rts['flux_error']).autopower(minimum_frequency=1/u.d,
                                                   maximum_frequency=5/u.d)
bfreq, bpower = LombScargle(bts.time, bts['flux'], bts['flux_error']).autopower(minimum_frequency=1/u.d,
                                                   maximum_frequency=5/u.d)

fig, ax = plt.subplots(3, figsize=[8,5], sharex=True, dpi=150)

ax[0].plot(gfreq, gpower, color='g', label='G')
ax[1].plot(rfreq, rpower, color='r', label='RP')
ax[2].plot(bfreq, bpower, color='b', label='BP')
ax[0].set_title('Lomb-Scargle Periodogram')
ax[2].set_xlabel('Frequency (Cycles per Day)')
ax[1].set_ylabel('Power')

for axis in ax.ravel():
    axis.tick_params(axis='both', which='major', labelsize=8)
    axis.legend(prop={'size': 8})
```

```python
gbestfreq = gfreq[np.argmax(gpower)]
rbestfreq = rfreq[np.argmax(rpower)]
bbestfreq = bfreq[np.argmax(bpower)]

print(f'G band period: {1/gbestfreq}')
print(f'RP band period: {1/rbestfreq}')
print(f'BP band period: {1/bbestfreq}')
```

We use double the period for folding because we are trying to fit a double Gaussian

```python
gts_folded = gts.fold(period=2./gbestfreq, normalize_phase=True, epoch_phase=0)
rts_folded = rts.fold(period=2./rbestfreq, normalize_phase=True, epoch_phase=0)
bts_folded = bts.fold(period=2./bbestfreq, normalize_phase=True, epoch_phase=0)
```

```python
fig = plt.figure(figsize=(6,3), dpi=150)

plt.errorbar(gts_folded['time'], gts_folded['flux'], yerr=gts_folded['flux_error'], fmt = 'g.')
plt.errorbar(rts_folded['time'], rts_folded['flux'], yerr=rts_folded['flux_error'], fmt = 'r.')
plt.errorbar(bts_folded['time'], bts_folded['flux'], yerr=bts_folded['flux_error'], fmt = 'b.')
plt.title('Light Curves Folded at Peak Frequency')
plt.xlabel('Phase')
plt.ylabel('Normalized Flux')
```

#### Fit a double Gaussian function

```python
from scipy.optimize import curve_fit
```

```python
def negDoubleGaussian(x, mu1, sig1, d1, mu2, sig2, d2, b):
    g1 = d1*np.exp(-np.power(x - mu1, 2.) / (2 * np.power(sig1, 2.)))
    g2 = d2*np.exp(-np.power(x - mu2, 2.) / (2 * np.power(sig2, 2.)))
    ndg = -g1 - g2 + b
    return ndg
```

In the double Gaussian function, `mu1` and `mu2` are the phase location of the dips, `sig1` and `sig2` are the widths, and `d1` and `d2` are the depths. `b` is the baseline flux, which we assume to be constant, but in reality may be better fit with a cosine function.

```python
p0 = np.array([-0.2, 0.2, 0.1, 0.3, 0.2, 0.1, 0.06]) # initial parameter estimates
pg, covg = curve_fit(negDoubleGaussian, gts_folded['time'], gts_folded['flux'], p0)
pr, covr = curve_fit(negDoubleGaussian, rts_folded['time'], rts_folded['flux'], p0)
pb, covb = curve_fit(negDoubleGaussian, bts_folded['time'], bts_folded['flux'], p0)
```

The `p` values are the best fit of the desired parameters, and `cov` is the covariance matrix, which can be used to extract the errors in the parameters.

```python
fig = plt.figure(figsize=(6,3), dpi=150)

plt.errorbar(gts_folded['time'], gts_folded['flux'], yerr=gts_folded['flux_error'], fmt = 'g.')
plt.errorbar(rts_folded['time'], rts_folded['flux'], yerr=rts_folded['flux_error'], fmt = 'r.')
plt.errorbar(bts_folded['time'], bts_folded['flux'], yerr=bts_folded['flux_error'], fmt = 'b.')
plt.title('Light Curves Folded at Peak Frequency')
plt.xlabel('Phase')
plt.ylabel('Normalized Flux')

xs = np.linspace(-0.5, 0.5, 100)
plt.plot(xs, negDoubleGaussian(xs, pg[0], pg[1], pg[2], pg[3], pg[4], pg[5], pg[6]), c='g')
plt.plot(xs, negDoubleGaussian(xs, pr[0], pr[1], pr[2], pr[3], pr[4], pr[5], pr[6]), c='r')
plt.plot(xs, negDoubleGaussian(xs, pb[0], pb[1], pb[2], pb[3], pb[4], pb[5], pb[6]), c='b')
```

We see that the model fits fairly well. The green light curve seems to be shifted slightly to the left of the blue and red ones; this is because the peak frequency of the G band periodogram is a tiny bit different from the other two. If we choose a common folding period, the three should appear more similar.

```python
gts_folded_common = gts.fold(period=2./gbestfreq, normalize_phase=True, epoch_phase=0)
rts_folded_common = rts.fold(period=2./gbestfreq, normalize_phase=True, epoch_phase=0)
bts_folded_common = bts.fold(period=2./gbestfreq, normalize_phase=True, epoch_phase=0)
```

```python
pg_common, covg_common = curve_fit(negDoubleGaussian, gts_folded_common['time'], gts_folded_common['flux'], p0)
pr_common, covr_common = curve_fit(negDoubleGaussian, rts_folded_common['time'], rts_folded_common['flux'], p0)
pb_common, covb_common = curve_fit(negDoubleGaussian, bts_folded_common['time'], bts_folded_common['flux'], p0)
```

```python
fig = plt.figure(figsize=(6,3), dpi=150)

plt.errorbar(gts_folded_common['time'], gts_folded_common['flux'], yerr=gts_folded_common['flux_error'], fmt = 'g.')
plt.errorbar(rts_folded_common['time'], rts_folded_common['flux'], yerr=rts_folded_common['flux_error'], fmt = 'r.')
plt.errorbar(bts_folded_common['time'], bts_folded_common['flux'], yerr=bts_folded_common['flux_error'], fmt = 'b.')
plt.title('Light Curves Folded at Common Peak Frequency')
plt.xlabel('Phase')
plt.ylabel('Normalized Flux')

xs = np.linspace(-0.5, 0.5, 100)
plt.plot(xs, negDoubleGaussian(xs, pg_common[0], pg_common[1], pg_common[2], pg_common[3], pg_common[4], pg_common[5], pg_common[6]), c='g')
plt.plot(xs, negDoubleGaussian(xs, pr_common[0], pr_common[1], pr_common[2], pr_common[3], pr_common[4], pr_common[5], pr_common[6]), c='r')
plt.plot(xs, negDoubleGaussian(xs, pb_common[0], pb_common[1], pb_common[2], pb_common[3], pb_common[4], pb_common[5], pb_common[6]), c='b')
```

The light curves clearly align much more closely when folded at a common peak frequency. From this, we can extract some information about the binary system, including the amplitude and phase of the transits, which can, for example, tell us about the relative sizes of the stars.
